# Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
```

```python
#focusing on the non float values 
TX_Housing_data = TX_Housing_data[["zipcode", 'population',
       'bedroom_number', 'bathroom_number', 'lining_space', 'land_space',
       'year_built', 'p_2015', 'p_2016', 'p_2017', 'p_2018', 'p_2019',
       'p_2020', 'p_2021', 'p_2022', 'p_2023', 'net_income' ]]

TX_Housing_data.head()
```

### Step 2: Split the data into X and y variables

```python
# Separate the data into labels and features

# Create a y variable and use the P_2023 column as the label
y = TX_Housing_data["p_2023"].ravel()


# Create an X variable and use all other columns as features
x = TX_Housing_data.drop(columns='p_2023')
```

```python
# Review the y variable series
y[:]
```

```python
# Review the X variable DataFrame
x.head(5)
```

```python
# Splitting into Train and Test sets
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(x_train)

# Scale the data
X_train_scaled = X_scaler.transform(x_train)
X_test_scaled = X_scaler.transform(x_test)
```

```python
x_train
```

```python
# Reset index 
x_train = x_train.reset_index(drop=True)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_features = len(x_train.columns)
layer_1 = 80
layer_2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_1,
                          input_dim=number_features,
                          activation='relu')
)

# Second hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_2,
                          input_dim=layer_1,
                          activation='relu')
)

# Output layer
nn.add(
    tf.keras.layers.Dense(units=1,
                          input_dim=layer_2,
                          activation='sigmoid')
)

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
