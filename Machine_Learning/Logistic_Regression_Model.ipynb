```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
df = pd.read_csv(
    Path(r"C:\Users\gurpa\Documents\GitHub\project-4-group-3\Resources\output\TX_full_data.csv")   
)

# Review the DataFrame
df.head()
```

#### Minor fixtures to dataset

```python
# Drop any null or empty values
df=df.dropna(how='any')
```

```python
#focusing on the relevant values 
df = df[["zipcode", "population","Net_Income", "P_2015", "P_2016","P_2017","P_2018","P_2019","P_2020","P_2021","P_2022","P_2023", "bedroom_number", "bathroom_number", 
                    "living_space", "land_space"]]

df.head()
```

### Step 2: Split the data into X and y variables

```python
# Separate the data into labels and features

# Create a y variable and use the P_2023 column as the label
y = df['P_2023'].values.reshape(-1,1)

# Create an X variable and use all other columns as features
X = df.drop(['P_2023'], axis = 1)
```

```python
# Review the y variable series
y[:]
```

```python
# Review the X variable DataFrame
X.head()
```

### Step 3: Create and Fit Model

```python
model = LogisticRegression()
```

```python
lab = preprocessing.LabelEncoder()
y_transformed = lab.fit_transform(y)
# Source https://www.statology.org/valueerror-unknown-label-type-continuous/
```

```python
model.fit(X, y_transformed)
```

```python
predicted_y_values = model.predict(X)
```

```python
# Create a copy of the original data
df_predicted = df.copy()

# Add a column with the predicted sales values
df_predicted["P_2024"] = predicted_y_values

# Display sample data
df_predicted.head()
```

### Step 4: Generate Results of Model

```python
# Compute the metrics for the linear regression model
score = model.score(X, y_transformed, sample_weight=None)
r2 = r2_score(y, predicted_y_values)
mse = mean_squared_error(y_transformed, predicted_y_values)
rmse = np.sqrt(mse)
std = np.std(y)

# Print relevant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is {std}.")
```
