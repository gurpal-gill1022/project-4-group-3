```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
df = pd.read_csv(
    Path(r"C:\Users\gurpa\Documents\GitHub\project-4-group-3\Resources\Datasets\TX_population_price_income.csv")   
)

# Review the DataFrame
df.head()
```

#### Minor fixtures to dataset

```python
# Drop any null or empty values
df=df.dropna(how='any')
```

```python
# Convert datatype of y data from float to int
df = df.astype({"P_2023":int},errors="raise")
```

### Step 2: Split the data into X and y variables, and then into training and testing sets

```python
# Separate the data into labels and features

# Create a y variable and use the P_2023 column as the label
y = df['P_2023'].values.reshape(-1,1)

# Create an X variable and use all other columns as features
X = df.drop(['P_2023', 'city', 'City', 'state', 'state_x'], axis = 1)
```

```python
# Review the y variable series
y[:]
```

```python
# Review the X variable DataFrame
X.head()
```

```python
# Split into testing and training sets using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Step 3: Fit a Logistic Regression Classifier

```python
lab = preprocessing.LabelEncoder()
y_transformed_train = lab.fit_transform(y_train)
```

```python
# Declare a logistic regression model.
# Apply a random_state of 9 to the model
logistic_regression_model = LogisticRegression(random_state=42)

# Fit and save the logistic regression model using the training data
lr_model = logistic_regression_model.fit(X_train, y_transformed_train)
```

### Step 4: Create predicted values for training and testing data

```python
#Generate training predictions
training_predictions = lr_model.predict(X_train)

#Generate testing predictions
testing_predictions = logistic_regression_model.predict(X_test)
```

### Step 5: Print confusion matric for the training data.

```python
# Create and save the confusion matrix for the training data
training_matrix = confusion_matrix(y_transformed_train, training_predictions)

# Print the confusion matrix for the training data
print(training_matrix)
```

### Step 6: Print confusion matric for the testing data.

```python
# Create and save the confusion matrix for the testing data
test_matrix = confusion_matrix(y_test, testing_predictions)

# Print the confusion matrix for the testing data
print(test_matrix)
```

### Step 7: Print the training classification report.

```python
# Create and save the training classification report
training_report = classification_report(y_transformed_train, training_predictions)

# Print the training classification report
print(training_report)
```

### Step 8: Print the testing classification report.

```python
# Create and save the testing classification report
testing_report = classification_report(y_test, testing_predictions)

# Print the testing classification report
print(testing_report)
```
