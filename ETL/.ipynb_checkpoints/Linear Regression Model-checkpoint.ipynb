# Linear Regression Model

```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Seeing the rows above and putting them to zero
TX_Housing_data.fillna(0, inplace=True)
```

```python
TX_Housing_data.info()
```

```python
TX_Housing_data.columns
```

```python
#focusing on the non float values 
TX_Housing_data = TX_Housing_data[["zipcode", 'population',
       'bedroom_number', 'bathroom_number', 'lining_space', 'land_space',
       'year_built', 'p_2015', 'p_2016', 'p_2017', 'p_2018', 'p_2019',
       'p_2020', 'p_2021', 'p_2022', 'p_2023', 'net_income' ]]

TX_Housing_data.head()
```

```python
# Create the X set by using the `reshape` function to format the ads data as a single column array.
X =  TX_Housing_data.drop(columns='p_2023')

# Display sample data
X[:5]
```

```python
y = TX_Housing_data["p_2023"]
y[:5]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
predicted_y_values = model.predict(X)
```

```python
# Create a copy of the original data
df_texas_predicted = TX_Housing_data.copy()

# Add a column with the predicted sales values
df_texas_predicted["p_2024"] = predicted_y_values

# Display sample data
df_texas_predicted.head()
```

```python
# Compute the metrics for the linear regression model
score = model.score(X, y, sample_weight=None)
r2 = r2_score(y, predicted_y_values)
mse = mean_squared_error(y, predicted_y_values)
rmse = np.sqrt(mse)
std = np.std(y)

# Print relevant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is {std}.")
```
